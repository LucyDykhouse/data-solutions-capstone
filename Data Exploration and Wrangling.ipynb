### Distribution of Numeric Variables

```python
fig, axes = plt.subplots(5, 2, figsize=(16,8), constrained_layout=True)
fig.suptitle('Distribution of Numeric Water Characteristics');

sns.histplot(ax=axes[0,0], data=df, x='ph');
sns.histplot(ax=axes[0,1], data=df, x='Hardness');
sns.histplot(ax=axes[1,0], data=df, x='Solids');
sns.histplot(ax=axes[1,1], data=df, x='Chloramines');
sns.histplot(ax=axes[2,0], data=df, x='Sulfate');
sns.histplot(ax=axes[2,1], data=df, x='Conductivity');
sns.histplot(ax=axes[3,0], data=df, x='Organic_carbon');
sns.histplot(ax=axes[3,1], data=df, x='Trihalomethanes');
sns.histplot(ax=axes[4,0], data=df, x='Turbidity');
sns.histplot(ax=axes[4,1], x=df['Potability']);
```

```python
normalized_df = (df - df.mean())/df.std()
```

```python
normalized_df.head(10)
```

```python
normalized_df.describe()
```

```python
sns.set(style='darkgrid')
sns.set_palette('tab10')

fig = sns.kdeplot(normalized_df['ph'])
fig = sns.kdeplot(normalized_df['Hardness'])
fig = sns.kdeplot(normalized_df['Solids'])
fig = sns.kdeplot(normalized_df['Chloramines'])
fig = sns.kdeplot(normalized_df['Sulfate'])
fig = sns.kdeplot(normalized_df['Conductivity'])
fig = sns.kdeplot(normalized_df['Organic_carbon'])
fig = sns.kdeplot(normalized_df['Trihalomethanes'])
fig = sns.kdeplot(normalized_df['Turbidity'])

plt.title('Distributions of Normalized Characteristics')
plt.xlabel('Normalized Value')
plt.legend(labels=['pH', 'Hardness', 'Solids', 'Chloramines', 'Sulfate', 'Conductivity', 'Organic_carbon', 'Trihalomethanes', 'Turbidity']);
```

### Examining Correlation between Variables

```python
potable_df = df[df['Potability'] == 1]
nonpotable_df = df[df['Potability'] == 0]
```

```python
potable_df.head()
```

```python
nonpotable_df.head()
```

```python
sns.set(style='darkgrid')

sns.histplot(data=nonpotable_df, x='ph', color='gold', label='Nonpotable', kde=True);
sns.histplot(data=potable_df, x='ph', color='teal', label='Potable', kde=True);

plt.title('Distribution of pH Based on Potability Status');
plt.xlabel('pH');
```

```python
sns.histplot(data=nonpotable_df, x='Hardness', color='gold', label='Nonpotable', kde=True);
sns.histplot(data=potable_df, x='Hardness', color='teal', label='Potable', kde=True);

plt.title('Distribution of Hardness Based on Potability Status');
plt.xlabel('Hardness');
```

```python
sns.histplot(data=nonpotable_df, x='Solids', color='gold', label='Nonpotable', kde=True);
sns.histplot(data=potable_df, x='Solids', color='teal', label='Potable', kde=True);

plt.title('Distribution of Solids Based on Potability Status');
plt.xlabel('Solids');
```

```python
sns.histplot(data=nonpotable_df, x='Chloramines', color='gold', label='Nonpotable', kde=True);
sns.histplot(data=potable_df, x='Chloramines', color='teal', label='Potable', kde=True);

plt.title('Distribution of Chloramines Based on Potability Status');
plt.xlabel('Chloramines');
```

```python
sns.histplot(data=nonpotable_df, x='Sulfate', color='gold', label='Nonpotable', kde=True);
sns.histplot(data=potable_df, x='Sulfate', color='teal', label='Potable', kde=True);

plt.title('Distribution of Sulfate Based on Potability Status');
plt.xlabel('Sulfate');
```

```python
sns.histplot(data=nonpotable_df, x='Conductivity', color='gold', label='Nonpotable', kde=True);
sns.histplot(data=potable_df, x='Conductivity', color='teal', label='Potable', kde=True);

plt.title('Distribution of Conductivity Based on Potability Status');
plt.xlabel('Conductivity');
```

```python
sns.histplot(data=nonpotable_df, x='Organic_carbon', color='gold', label='Nonpotable', kde=True);
sns.histplot(data=potable_df, x='Organic_carbon', color='teal', label='Potable', kde=True);

plt.title('Distribution of Total Organic Carbon Based on Potability Status');
plt.xlabel('Organic Carbon');
```
